```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
mydata = pd.read_csv('/content/creditcard_2023.csv')
```

```python
mydata.head()
```

```python
#Determing number of rows and column
num_rows = mydata.shape[0]
num_columns = mydata.shape[1]

#Print the number of rows and columns
print(f"Number of rows: {num_rows }")
print (f"Number of columns: {num_columns}")
```

```python
mydata.dtypes
```

```python
#Data Cleaning

#Checking for missing values
mydata.isnull().sum()
```

```python
#Handling missing values by imputation
mydata['V14'].fillna(mydata['V14'].mean(),inplace=True)
mydata['V15'].fillna(mydata['V15'].mean(),inplace=True)
mydata['V16'].fillna(mydata['V16'].mean(),inplace=True)
mydata['V17'].fillna(mydata['V17'].mean(),inplace=True)
mydata['V18'].fillna(mydata['V18'].mean(),inplace=True)
mydata['V19'].fillna(mydata['V19'].mean(),inplace=True)
mydata['V20'].fillna(mydata['V20'].mean(),inplace=True)
mydata['V21'].fillna(mydata['V21'].mean(),inplace=True)
mydata['V22'].fillna(mydata['V22'].mean(),inplace=True)
mydata['V23'].fillna(mydata['V23'].mean(),inplace=True)
mydata['V24'].fillna(mydata['V24'].mean(),inplace=True)
mydata['V25'].fillna(mydata['V19'].mean(),inplace=True)
mydata['V26'].fillna(mydata['V19'].mean(),inplace=True)
mydata['V27'].fillna(mydata['V19'].mean(),inplace=True)
mydata['V28'].fillna(mydata['V19'].mean(),inplace=True)
mydata['Amount'].fillna(mydata['V19'].mean(),inplace=True)
mydata['Class'].fillna(mydata['V19'].mean(),inplace=True)
```

```python
mydata.isnull().sum()
```

```python
#Visualization

#4.Histogram
plt.hist(mydata['V1'])
plt.xlabel('V1')
plt.ylabel('Frequency')
plt.title('Distribution of V1')
plt.show()

```

```python
#5.Calculate the skewness
skewness_V2  = mydata['V2'].skew()
skewness_V3 = mydata['V3'].skew()

print(skewness_V2)
print(skewness_V3)
print()

#Visualize skewness
plt.figure(figsize = (10,6))
plt.subplot(2,1,1)
plt.hist(mydata['V2'], color = 'pink', alpha = 0.5)
plt.title( 'Distribution of V2')
plt.xlabel('Values of V2')
plt.ylabel('Frequency')

plt.subplot(2,1,2)
plt.hist(mydata['V3'], color = 'blue', alpha = 0.5)
plt.title( 'Distribution of V3')
plt.xlabel('Values of V3')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show


```

```python
#6.Bar plot for orders

#Group the data by category and calculate the number of orders
category_orders = mydata.groupby('V4')['id'].count()

#Create a bar plot
plt.bar(category_orders.index, category_orders.values)
plt.xlabel('V4')
plt.ylabel('Total Number of Orders')
plt.title('Total Number of Orders by category')
plt.show()
```

```python
#7. Line Plot of amount over time
#plot the data
plt.plot(mydata['Class'],mydata['Amount'])
plt.title('Trend in Amount over Class')
plt.xlabel('Class')
plt.ylabel('Amount')
plt.show()

```

```python
#8 Create a scatter plot
plt.scatter(mydata['id'],mydata['V5'])
plt.title('Scatter plot id vs V5')
plt.xlabel('id')
plt.ylabel('V5')
plt.show()
```

```python


#Create a box plot
plt.boxplot(mydata['V6'])
plt.xlabel('class')
plt.ylabel('V6')
plt.show
```
